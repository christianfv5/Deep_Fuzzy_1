```python
import re
import matplotlib.pyplot as plt


def read_history(filename):
    history = {
        "loss": [],
        "acc": [],
        "val_loss": [],
        "val_acc": [],
    }

    with open(filename, "r") as infile:
        for line in infile:
            line = line.strip()
            idx = line.find("- loss:")

            if idx == -1:
                continue

            data = re.split(" ?\- [a-z_]+: ", line[idx:])[1:]
            loss, acc, val_loss, val_acc = map(lambda x: float(x), data)

            history["loss"].append(loss)
            history["acc"].append(acc)
            history["val_loss"].append(val_loss)
            history["val_acc"].append(val_acc)
            
    return history
```

```python
lenet_history = read_history("training_history/lenet.txt")
anfis_lenet_history = read_history("training_history/anfis_lenet.txt")


plt.figure(figsize=(8,6))

ax1, = plt.plot(lenet_history["acc"], "g--")
ax2, = plt.plot(lenet_history["val_acc"], "g")
ax3, = plt.plot(anfis_lenet_history["acc"], "r--")
ax4, = plt.plot(anfis_lenet_history["val_acc"], "r")

plt.title("CIFAR-10 Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")
plt.ylim(0.65, 0.8)

plt.legend(
    [ax1, ax2, ax3, ax4],
    ["LeNet Training Set", 
     "LeNet Validation Set",
     "DNFIS Training Set",
     "DNFIS Validation Set"])

plt.savefig("figures/lenet_cifar_acc.pdf")


plt.figure(figsize=(8,6))

ax1, = plt.plot(lenet_history["loss"], "g--")
ax2, = plt.plot(lenet_history["val_loss"], "g")
ax3, = plt.plot(anfis_lenet_history["loss"], "r--")
ax4, = plt.plot(anfis_lenet_history["val_loss"], "r")

plt.title("CIFAR-10 Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.legend(
    [ax1, ax2, ax3, ax4],
    ["LeNet Training Set", 
     "LeNet Validation Set",
     "DNFIS Training Set",
     "DNFIS Validation Set"])

plt.savefig("figures/lenet_cifar_loss.pdf")
```
